## Matplotlib example workflow

```python
# 0. import matplotlin and get ready to plot in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2 , 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")


# 5. Save & show the figure
fig.savefig("../images/sample-plot.png")
```

## 
Making figures with Numpy arrays

We want:
- Line plot
- Scatter plot
- Bar plot
- Histograms
- Subplots

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100) # returns evenly space numbers
x
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use some data to build some scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.tan(x));
```

```python
# Make a plot from Dictionary
nut_butter_prices = {"Almond Butter" : 10,
                     "Peanut Butter" : 8,
                     "Cashew Butter" : 12}
fig, ax = plt.subplots();
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title = "Dan's Nut Butter Store", ylabel = "Price ($)")

fig.savefig("../images/butter-barplot.png")
```

### barh() requires x-y coordinates in list

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title = "Dan's Nut Butter Store", xlabel = "Price ($)");
```

```python
# Make some data and plot histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
```

```python
# let's plot sin, cos, tan, cosec, sec, cot 
x = np.linspace(-np.pi, np.pi, 100)
x


fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(nrows = 3,
                                                          ncols = 2,
                                                          figsize = (12, 12))

ax1.plot(x, np.sin(x))
ax1.set(title = "sin(x)")
ax2.plot(x, np.cos(x))
ax2.set(title = "cos(x)")
ax3.plot(x, np.tan(x))
ax3.set(title = "tan(x)")
ax4.plot(x, np.arcsin(x))
ax4.set(title = "sin-1(x)")
ax5.plot(x, np.arccos(x))
ax5.set(title = "cos-1(x)")
ax6.plot(x, np.arctan(x));
ax6.set(title = "tan-1(x)")

fig.savefig("../images/trigo-subplot.png")
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                                          ncols = 2,
                                                          figsize = (10, 8))
ax1.plot(x, x/2);
ax2.scatter(np.random.randn(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2

fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 8))

ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.randn(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
