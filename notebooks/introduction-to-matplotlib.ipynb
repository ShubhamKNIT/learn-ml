## Matplotlib example workflow

```python
# 0. import matplotlin and get ready to plot in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2 , 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")


# 5. Save & show the figure
fig.savefig("../images/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))
```

```python
# let's plot sin, cos, tan, cosec, sec, cot 
x = np.linspace(-np.pi, np.pi, 100)
x


fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(nrows = 3,
                                                          ncols = 2,
                                                          figsize = (12, 12))

ax1.plot(x, np.sin(x))
ax1.set(title = "sin(x)")
ax2.plot(x, np.cos(x))
ax2.set(title = "cos(x)")
ax3.plot(x, np.tan(x))
ax3.set(title = "tan(x)")
ax4.plot(x, np.arcsin(x))
ax4.set(title = "sin-1(x)")
ax5.plot(x, np.arccos(x))
ax5.set(title = "cos-1(x)")
ax6.plot(x, np.arctan(x));
ax6.set(title = "tan-1(x)")

fig.savefig("../images/trigo-subplot.png")
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                                          ncols = 2,
                                                          figsize = (10, 8))
ax1.plot(x, x/2);
ax2.scatter(np.random.randn(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

## Which one should I use? (pyplot vs matplotlib OO method?)

- When plotting quickly, its okay to use Pyplot methods
- When plotting something more advanced, use the OO methods

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# using pyplot method
over_50.plot(kind = "scatter",
             x = "age", 
             y = "chol",
             c = "target",
             figsize = (10, 6))
```

```python
# OO Method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = "scatter",
             x = "age", 
             y = "chol",
             c = "target", 
             ax = ax);

# ax.set_xlim(45, 100);
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10 ,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"], 
                     c = over_50["target"])

# customize the plot
ax.set(title = "Heart Disease and Cholestrol Levels",
       xlabel = "age", 
       ylabel = "Cholestrol");

# add a legend (*scatter to upack elements)
ax.legend(*scatter.legend_elements(), title = "Target");

# adding horizontal line
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (12, 12))

# add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Customize ax0
ax0.set(title = "Heart-Disease and Cholestoral Levels",
        xlabel = "Age",
        ylabel = "Cholestoral")

# adding legends to ax0
ax0.legend(*scatter.legend_elements(),
           title = "Target");

# add a meanline
ax0.axhline(over_50["chol"].mean(),
            linestyle = "--");


# plotting another subplot ax1

# add data to ax1
scatter = ax1.scatter(x = over_50["age"], 
                      y = over_50["thalach"],
                      c = over_50["target"],)

# Customize ax1
# thalach = maximum heart rate achived
ax1.set(title = "Heart-Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")

# adding a horizontal line
ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

# ading tiltle to figure
fig.suptitle("Heart diseasea anlysis", fontsize = 30, fontweight = "bold");
```

```python
# set the style
plt.style.use("seaborn-v0_8-whitegrid")

# OO method from scratch
fig, ax = plt.subplots(figsize = (10 ,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"], 
                     c = over_50["target"],
                     cmap = "winter") 

# customize the plot
ax.set(title = "Heart Disease and Cholestrol Levels",
       xlabel = "age", 
       ylabel = "Cholestrol");

# add a legend (*scatter to upack elements)
ax.legend(*scatter.legend_elements(), title = "Target");

# adding horizontal line
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
# set the style
plt.style.use("seaborn-v0_8-whitegrid")

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (12, 12))

# add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                      cmap = "winter")

# Customize ax0
ax0.set(title = "Heart-Disease and Cholestoral Levels",
        xlabel = "Age",
        ylabel = "Cholestoral")

# changing xlim and ylim
ax0.set_xlim([50, 80])
ax0.set_ylim([100, 600])

# adding legends to ax0
ax0.legend(*scatter.legend_elements(),
           title = "Target");

# add a meanline
ax0.axhline(over_50["chol"].mean(),
            linestyle = "--");


# plotting another subplot ax1

# add data to ax1
scatter = ax1.scatter(x = over_50["age"], 
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = "winter")

# Customize ax1
# thalach = maximum heart rate achived
ax1.set(title = "Heart-Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# changing xlim and ylim
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# adding a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")

# adding a horizontal line
ax1.axhline(over_50["thalach"].mean(), linestyle = "--")

# ading tiltle to figure
fig.suptitle("Heart diseasea anlysis", fontsize = 30, fontweight = "bold");

# saving images
fig.savefig("../images/heart-disease-analysis-plot.png")
```
