## Probability Distribution of Tossing coins thrice

```python
import numpy as np

N = 500000  # total iterations
M = 3  # total coin flips

heads = np.zeros(M + 1)  # count of heads

for i in range(N):
    flips = np.random.randint(0, 2, M)  # simulate coin flips
    h, _ = np.bincount(flips, minlength=2)  # count heads and tails
    heads[h] += 1  # update heads count

prob = heads/N  # calculate probabilities
print("Probabilites: %s" % np.array2string(prob))

import matplotlib.pyplot as plt
plt.bar(range(M + 1), prob)  # plot probabilities
```

## Probaility of Sharing Birthday Date

```python
p = 1 - (1/365)  # Probability that the next person has a different birthday
n = 0
total_p = 1  # Start with probability of no shared birthdays

while total_p > 0.5:
    n += 1
    total_p *= (365 - n + 1) / 365  # Update probability

print("Total number of people required to share a birthday date: " + str(n))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize arrays for matches and probabilities
max_people = 32
matches = np.zeros(max_people, dtype=np.int32)
prob = np.zeros(max_people)

# Simulate birthday problem for different group sizes
num_simulations = 100000  # Number of simulations
for m in range(1, max_people):
    matches[m] = 0
    for n in range(num_simulations):
        b = np.random.randint(0, 365, m)  # Random birthdays
        if len(b) != len(set(b)):  # Check for duplicates
            matches[m] += 1  # Count if there's at least one match
    prob[m] = matches[m] / num_simulations  # Calculate probability
    print("%2d %0.6f" % (m, prob[m]))  # Print results

# Create scatter and line plots
plt.scatter(range(1, max_people), prob[1:], color='blue', label='Simulated Probability')  # Scatter plot
plt.plot(range(1, max_people), prob[1:], color='red')  # Line plot
plt.axhline(y=0.5, color='green', linestyle='--', label='50% Probability')  # Horizontal line at y=0.5

# Add titles and labels
plt.title("Probability Distribution of Shared Birthdays")
plt.xlabel("Number of People")
plt.ylabel("Probability of a Shared Birthday")
plt.legend()
plt.grid()
plt.show()  # Display the plot
```

# Key Probability Formulas

1. **Conditional Probability**:
  $$ P(A | B) = \frac{P(A \cap B)}{P(B)} \quad \text{if } P(B) > 0 $$

3. **Joint Probability**:
   $$ P(A \cap B) = P(A | B) \cdot P(B) = P(B | A) \cdot P(A) $$

4. **Marginal Probability**:
   $$ P(A) = \sum_{B} P(A \cap B) \quad \text{(for discrete events)} $$

5. **Bayes' Theorem**:
    $$ P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)} $$

6. **Chain Rule of Probability**:
    $$ P(A_1, A_2, \ldots, A_n) = P(A_1) \cdot P(A_2 | A_1) \cdot P(A_3 | A_1, A_2) \cdots P(A_n | A_1, A_2, \ldots, A_{n-1}) $$

## Summary
- **Conditional Probability**: Calculates the probability of one event given another.
- **Joint Probability**: Measures the probability of two events happening together.
- **Marginal Probability**: Gives the overall probability of a single event.
- **Bayes' Theorem**: Updates the probability of a hypothesis based on new evidence.
- **Chain Rule**: Expresses the joint probability of multiple events as a product of conditional probabilities.
